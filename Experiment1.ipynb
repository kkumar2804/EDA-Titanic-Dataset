```python
# Import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Count 'pclass' values
data['pclass'].value_counts()
```

```python
print(np.mean(data['survived'][data['pclass'] == 1]) * 100)  # Class 1 survival rate
print(np.mean(data['survived'][data['pclass'] == 2]) * 100)  # Class 2 survival rate
print(np.mean(data['survived'][data['pclass'] == 3]) * 100)  # Class 3 survival rate
```

```python
data[['pclass', 'survived']].groupby(['pclass']).mean() * 100  # Mean survival by class
```

```python
data['sex'].value_counts()  # Count 'sex' values
```

```python
print(np.mean(data['survived'][data['sex'] == 'male']) * 100)  # Male survival rate
print(np.mean(data['survived'][data['sex'] == 'female']) * 100)  # Female survival rate
```

```python
data[['sex', 'survived']].groupby(['sex']).mean() * 100  # Mean survival by gender
```

```python
print(len(data))  # Total number of rows
```

```python
print(len(data[data['survived'] == 1]))  # Total survivors
```

```python
sns.catplot(x = 'sex', hue = 'survived', kind = 'count' , data = data)  # Count plot: sex vs survived
```

```python
sns.catplot(x= 'sex', y = 'survived', hue = 'sex', kind = 'bar', data = data)  # Bar plot: sex vs survived
```

```python
sns.catplot(x = 'sex', y = 'survived', hue = 'pclass', kind = 'violin', data = data)  # Violin plot: sex vs survived by pclass
```

```python
sns.catplot(x = 'sex', y = 'survived', hue = 'pclass', kind = 'bar', data = data)  # Bar plot: sex vs survived by pclass
```

```python
sns.catplot(x = 'sex', hue = 'pclass', kind = 'count', data = data)  # Count plot: sex by pclass
```

```python
data[data['sex'] == 'male'].groupby('pclass')['survived'].sum()  # Male survivors by pclass
```

```python
sns.catplot(x = "pclass", y = 'survived', hue = 'sex', kind = 'bar', data = data)  # Bar plot: pclass vs survived by sex
```

```python
# Count non-survived females by pclass
data[
    (data['sex'] == 'female') & (data['survived'] == 0)
].groupby('pclass')['survived'].count()
```

```python
# Count non-survived males by pclass
data[
    (data['sex'] == 'male') & (data['survived'] == 0)
].groupby('pclass')['survived'].count()
```

```python
# Count survived females by pclass
data[
    (data['sex'] == 'female') & (data['survived'] == 1)
].groupby('pclass')['survived'].count()
```

```python
# Count survived males by pclass
data[
    (data['sex'] == 'male') & (data['survived'] == 1)
].groupby('pclass')['survived'].count()
```

```python
sns.heatmap(data.isnull(), cmap = 'magma')  # Heatmap of missing values
```
